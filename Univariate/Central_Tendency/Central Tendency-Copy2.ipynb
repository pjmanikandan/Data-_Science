
# IQR Interquartile Range


```python
univariate=pd.DataFrame(index=["Mean","Median","Mode","Q1:25%","Q2:50%","Q3:75%",
                               "99%","Q4:100%","IOR","1.5rule","Lesser","Greater","Min","Max"],columns=quan)
for columnName in quan:
    univariate[columnName]["Mean"]=dataset[columnName].mean()
    univariate[columnName]["Median"]=dataset[columnName].median()
    univariate[columnName]["Mode"]=dataset[columnName].mode()[0]
    univariate[columnName]["Q1:25%"]=dataset.describe()[columnName]["25%"]
    univariate[columnName]["Q2:50%"]=dataset.describe()[columnName]["50%"]
    univariate[columnName]["Q3:75%"]=dataset.describe()[columnName]["75%"]
    univariate[columnName]["99%"]=np.percentile(dataset[columnName],99)
    univariate[columnName]["Q4:100%"]=dataset.describe()[columnName]["max"]
    univariate[columnName]["IOR"]=univariate[columnName]["Q3:75%"]-univariate[columnName]["Q1:25%"]
    univariate[columnName]["1.5rule"]=1.5*univariate[columnName]["IOR"]
    univariate[columnName]["Lesser"]=univariate[columnName]["Q1:25%"]-univariate[columnName]["1.5rule"]
    univariate[columnName]["Greater"]=univariate[columnName]["Q3:75%"]+univariate[columnName]["1.5rule"]
    univariate[columnName]["Min"]=dataset[columnName].min()
    univariate[columnName]["Max"]=dataset[columnName].max()
    
```

```python
univariate
```

```python
lesser=[]
greater=[]

for columnName in quan:
    if(univariate[columnName]["Min"]<univariate[columnName]["Lesser"]):
        lesser.append(columnName)
    if(univariate[columnName]["Max"]>univariate[columnName]["Greater"]):
        greater.append(columnName)
    
    
```

```python
greater
```
